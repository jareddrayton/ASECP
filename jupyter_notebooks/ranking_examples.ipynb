```python
import math
import random
import matplotlib.pyplot as plt

individuals = 1000

# Exponential ranking using Eiben

probabilities = []

for i in range(1, individuals+1):
    prob = (1.0 - (math.e ** - i)) / (individuals -1)

    probabilities.append(prob)

#print(probabilities)
print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()
```

```python
# Linear rankign as given in Eiben

selection_pressure = 2.0  # This can be in the range 1.0 < x <= 2.0

s = 1.5
probabilities = []

for i in range(individuals):
    prob = (2 - s) / individuals + (2.0 * i * (s - 1)) / (individuals * (individuals - 1))

    probabilities.append(prob)
    #print(i, prob)

print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()
    
```

```python
# Exponential ranking as given in ``A Comparison of Selection Schemes used in Genetic Algorithms pg 34

c = 0.95

individuals = 150

probabilities = []

for i in range(1, individuals+1):
    prob = ((c - 1) / ((c ** individuals) - 1)) * (c ** (individuals - i))

    probabilities.append(prob)

print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()
```

```python
# Geometric ranking as given in Evolutionary computation Volume 1

probabilities = []

individuals = 100

A = 0.5 # 0.0 < A < 1.0

for i in range(individuals):
    prob = A * (1-A) ** (individuals-1-i)

    probabilities.append(prob)



print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()

```
