```python
# Square ranking as given in Evolutionary computation Volume 1 page 188

B = 2 # 1 - 2
A = 2 - B

individuals = 10

c_control = (B-A) * individuals * (2 * individuals - 1) / ((6 * (individuals - 1)) + (individuals * A))  

c_test = individuals * (2 * individuals - 1) / (6 * (individuals - 1))

print(A, B)

print("Normalisation factor", c_control)
print("Normalisation factor", c_test)

prob_control = []
prob_test = []


for i in range(1, individuals+1):

    prob_c = A + ((i ** 2)/(individuals - 1)**2) * (B - A)  / c_control

    prob_control.append(prob_c)
    
    prob_t = ((i ** 2)/(individuals - 1)**2)   / c_test
    
    prob_test.append(prob_t)
    
    
print(prob_control)
print(sum(prob_control))

print(prob_test)
print(sum(prob_test))



fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4.5))

ax1.plot(prob_control)
ax2.plot(prob_test)
plt.show()
```

$$ Pr_{sqrank}(i) = \frac{rank(i)^{2} / (\mu - 1)^{2}} {c} $$

$$ c = \frac{\mu(2\mu - 1)}{6(\mu - 1)} $$
