```python
import math
import random
import matplotlib.pyplot as plt

individuals = 50

# Exponential ranking using Eiben

probabilities = []

for i in range(1, individuals+1):
    prob = (1.0 - (math.e ** - i)) / (individuals -1)

    probabilities.append(prob)

#print(probabilities)
print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()
```

```python
# Exponential ranking as given in ``A Comparison of Selection Schemes used in Genetic Algorithms pg 34

c = 0.98

individuals = 100

probabilities = []

for i in range(1, individuals+1):
    prob = ((c - 1) / ((c ** individuals) - 1)) * (c ** (individuals - i))

    probabilities.append(prob)

print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()
```

```python
# Square ranking as given in Evolutionary computation Volume 1 page 188

B = 1.9 # 1 - 2
A = 0

individuals = 50

c = ((B-A) * individuals * ((2 * individuals)-1)) / ((6 * (individuals - 1)) + (individuals * A))  #0.5 

print(c)

probabilities = []



for i in range(individuals):
    prob = ((A + abs((i ** 2)/(individuals - 1)**2)) * (B - A))  / c

    probabilities.append(prob)

print(probabilities)
print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()
```

```python
# Geometric ranking as given in Evolutionary computation Volume 1

probabilities = []

A = 0.8 # 0.0 < A < 1.0

for i in range(individuals):
    prob = A * (1-A) ** (individuals-1-i)

    probabilities.append(prob)



print(sum(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()

```
