```python
import math
import random
import matplotlib.pyplot as plt



individuals = 50



# Exponential ranking

probabilities = []

for i in range(1, individuals+1):
    prob = (1.0 - (math.e ** - i)) / individuals

    probabilities.append(prob)
    
print(probabilities)
print(sum(probabilities))
print(len(probabilities))

fig, ax = plt.subplots()
ax.plot(probabilities)
plt.show()




```
